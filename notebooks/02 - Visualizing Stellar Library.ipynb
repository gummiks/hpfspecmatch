# Visualizing Stellar Library

## Import packages

```python
from importlib import reload
import os
import glob2
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import hpfspec
import hpfspecmatch
```

## Read in Library 

```python
# download library if does not exist
hpfspecmatch.utils.get_library()
```

```python
# Load list of fits files
LIBRARY_DIR = '../library/20201008_specmatch_nir/'
library_fitsfiles = glob2.glob(LIBRARY_DIR+'FITS/*/*.fits')
```

```python
# Read in all files as an HPFSpecList object
HLS = hpfspec.HPFSpecList(filelist=library_fitsfiles)
```

```python
# List of all of the HPF spectra
HLS.splist
```

```python
# More info on the targets
HLS.df
```

```python
# Read in required information on all of the 
# This has the Teff, Fe/H, and logg for all of the stars
# OBJECT_ID is the HPF name of the star
df_lib = pd.read_csv(LIBRARY_DIR+'20201008_specmatch_nir.csv')
df_lib
```

## Plot Library

```python
# PlotS Teff vs Fe/H and Teff vs logg for all library stars

fig = plt.figure(figsize=(18,10))
gs = gridspec.GridSpec(1, 2)
ax = plt.subplot(gs[0, 0])
bx = plt.subplot(gs[0, 1])

ax.set_ylabel('Teff [K]',fontsize=15)
ax.set_xlabel('Fe/H',fontsize=15)
bx.set_xlabel('log(g)',fontsize=15)
fig.suptitle('Stellar Library Parameters',fontsize=19)

ax.errorbar(df_lib['[Fe/H]'], df_lib['Teff'],df_lib['e_Teff'],df_lib['e_[Fe/H]'],lw=0,elinewidth=1.,marker="o",mew=0.5,markersize=8,capsize=4)
bx.errorbar(df_lib['log(g)'], df_lib['Teff'],df_lib['e_Teff'],df_lib['e_log(g)'],lw=0,elinewidth=1.,marker="o",mew=0.5,markersize=8,capsize=4)

plt.show()
```
